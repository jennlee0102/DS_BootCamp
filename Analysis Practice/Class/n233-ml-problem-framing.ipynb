## ÌèâÍ∞ÄÏßÄÌëú ÏÑ§Ï†ï
- Î¨∏Ï†úÏùò Ïú†ÌòïÏóê ÎßûÎäî ÌèâÍ∞ÄÏßÄÌëúÎ•º ÏÑ†ÌÉùÌïòÏó¨ Î™®Îç∏Ïùò ÏÑ±Îä•ÏùÑ ÌèâÍ∞ÄÌï¥Ïïº Ìï©ÎãàÎã§.
  - Î∂ÑÎ•ò ÌèâÍ∞ÄÏßÄÌëúÏôÄ ÌöåÍ∑Ä ÌèâÍ∞ÄÏßÄÌëúÎ•º ÌòºÎèôÌï¥ÏÑ† Ïïà Îê©ÎãàÎã§.
  - ÌïòÎÇòÏùò ÌèâÍ∞ÄÏßÄÌëúÎßåÏùÑ ÏÇ¨Ïö©Ìï¥ Î™®Îç∏ ÏÑ±Îä•ÏùÑ Í≤ÄÏ¶ùÌïòÎäî Í≤ΩÏö∞Îäî Í±∞Ïùò ÏóÜÏäµÎãàÎã§. Ïó¨Îü¨ ÌèâÍ∞ÄÏßÄÌëúÎ•º Ìï®Íªò ÏÇ¥Ìé¥Î≥¥Î©∞ Î™®Îç∏Ïùò ÏòàÏ∏° ÏñëÏÉÅÏùÑ ÌååÏïÖÌï©ÎãàÎã§.
<br></br>

- ÎåÄÌëúÏ†ÅÏù∏ ÌöåÍ∑Ä ÌèâÍ∞ÄÏßÄÌëúÎäî ÏïÑÎûòÏôÄ Í∞ôÏäµÎãàÎã§.
  - $MSE$: ÎØ∏Î∂Ñ Í∞ÄÎä•ÌïòÏó¨ ÎπÑÏö© Ìï®ÏàòÎ°ú Í∞ÄÏû• ÌùîÌïòÍ≤å ÏÇ¨Ïö©Îê©ÎãàÎã§. ÌïòÏßÄÎßå Ï†úÍ≥±ÏùÑ Ï∑®ÌïòÏó¨ Ïù¥ÏÉÅÏπòÏóê ÎØºÍ∞êÌï©ÎãàÎã§.
  - $RMSE$: $MSE$Ïóê Î£®Ìä∏Î•º ÏîåÏõå, Ïù¥ÏÉÅÏπòÏóê ÏÉÅÎåÄÏ†ÅÏúºÎ°ú Îçú ÎØºÍ∞êÌï©ÎãàÎã§.
  - $MAE$: Ï†àÎåÄÏò§Ï∞®Ïùò ÌèâÍ∑†ÏúºÎ°ú, ÏßÅÍ¥ÄÏ†ÅÏúºÎ°ú Ïù¥Ìï¥ÌïòÍ∏∞ Ïâ¨Ïõå ÏÑ†Ìò∏Îê©ÎãàÎã§.
  - $R^2$: Î™®Îç∏Ïù¥ Îç∞Ïù¥ÌÑ∞Îì§ÏùÑ ÏñºÎßàÎÇò Ïûò ÏÑ§Î™ÖÌïòÎäîÏßÄÎ•º Î≥¥Ïó¨Ï§çÎãàÎã§.
<br></br>

- ÎåÄÌëúÏ†ÅÏù∏ Î∂ÑÎ•ò ÌèâÍ∞ÄÏßÄÌëúÎäî ÏïÑÎûòÏôÄ Í∞ôÏäµÎãàÎã§.
  - $Accuracy=\frac{TP+TN}{TP+TN+FP+FN}$
  - $Precision=\frac{TP}{TP+FP}$
  - $Recall=\frac{TP}{TP+FN}$
  - $F1=\frac{2}{1/Precision + 1/Recall}$
  - $ROC-AUC$

  - ÌÉÄÍ≤üÎ∂ÑÌè¨Í∞Ä Î∂àÍ∑†ÌòïÌï† ÎïåÎäî accuracyÎäî Ï†ÅÏ†àÌïú ÌèâÍ∞ÄÏßÄÌëúÍ∞Ä ÏïÑÎãôÎãàÎã§. Ïù¥ Ïô∏Ïóê Îã§Î•∏ ÌèâÍ∞ÄÏßÄÌëúÎ°ú Î™®Îç∏Ïùò ÏÑ±Îä•ÏùÑ ÌèâÍ∞ÄÌï¥ÏïºÌï©ÎãàÎã§.

### Î∂ÑÎ•ò Î¨∏Ï†úÏóêÏÑúÎäî Confusion MatrixÎ•º Í≥ÑÏÇ∞ÌïòÏó¨ ÏßÅÍ¥ÄÏ†ÅÏúºÎ°ú Î™®Îç∏Ïùò ÏòàÏ∏° Í≤∞Í≥ºÎ•º ÌôïÏù∏Ìï† Ïàò ÏûàÏäµÎãàÎã§.

```python
from sklearn.metrics import plot_confusion_matrix

plot_confusion_matrix(pipe, X_test, y_test, cmap=plt.cm.Reds)
plt.title(f"Confusion Matrix (n = {len(y_test)})", fontsize=15)
```

### Classification ReportÎ•º ÌÜµÌï¥ Í∞Å classÏóê ÎåÄÌïú ÏÑ±Îä• ÌèâÍ∞Ä Í≤∞Í≥ºÎ•º ÌôïÏù∏Ìï† Ïàò ÏûàÏäµÎãàÎã§.

```python
from sklearn.metrics import classification_report

y_pred = pipe.predict(X_test)
print(classification_report(y_test, y_pred))
```

### ROC Í≥°ÏÑ†Í≥º AUC Ïä§ÏΩîÏñ¥Î•º ÌÜµÌï¥ Î™®Îç∏Ïù¥ ÏòàÏ∏°Ìïú Í∞í(ÌôïÎ•†, Logit Îì±)Ïóê ÎåÄÌïú ÏÑ±Îä•ÏùÑ ÌôïÏù∏Ìï† Ïàò ÏûàÏäµÎãàÎã§.
- ROC Í≥°ÏÑ†ÏùÄ FPR Î≥ÄÌôîÏóê Îî∞Î•∏ TPRÏùò Î≥ÄÌôîÏùò ÏûêÏ∑®Î•º Ï∞çÏùÄ Í∑∏ÎûòÌîÑÏûÖÎãàÎã§.
- AUC Ïä§ÏΩîÏñ¥Îäî ROC Í≥°ÏÑ† ÏïÑÎû´Î∂ÄÎ∂ÑÏùò Î©¥Ï†ÅÏù¥Î©∞, Í∞íÏù¥ ÌÅ¥ ÏàòÎ°ù Î™®Îç∏ÏùÄ True Îç∞Ïù¥ÌÑ∞ÏôÄ False Îç∞Ïù¥ÌÑ∞Î•º Ïûò Íµ¨Î∂ÑÌï©ÎãàÎã§.

```python
from sklearn.metrics import roc_curve, roc_auc_score

y_pred_proba = pipe.predict_proba(X_test)[:, 1]
print("AUC:", roc_auc_score(y_test, y_pred_proba))

baseline_fpr, baseline_tpr, _ = roc_curve(y_test, [0] * len(y_test))
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

plt.plot(baseline_fpr, baseline_tpr, linestyle="--", color="green", label="Baseline")
plt.plot(fpr, tpr, color="blue", label="Model")
plt.scatter(fpr, tpr, color="red")
plt.legend()
plt.title("ROC curve")
plt.xlabel("FPR")
plt.ylabel("TPR")
```

## ÌÉÄÍ≤üÏùò Î∂ÑÌè¨Í∞Ä Î∂àÍ∑†ÌòïÏùº Îïå

### üè† Case Study: House Price Prediction

```python
df = pd.read_csv(
    "https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/house-prices/house_prices_train.csv"
)
```

```python
# ÏòàÏ∏°ÌïòÍ≥†Ïûê ÌïòÎäî Ïª¨ÎüºÏùÄ SalePriceÏûÖÎãàÎã§. Î∂ÑÌè¨Î•º ÌôïÏù∏Ìï¥ Î≥ºÍπåÏöî?
features = [
    "BldgType",
    "CentralAir",
    "FullBath",
    "GrLivArea",
    "LotArea",
    "OverallCond",
    "OverallQual",
    "TotRmsAbvGrd",
    "YearBuilt",
    "YrSold",
]
target = "SalePrice"
sns.displot(df[target])
```

- ÌÉÄÍ≤ü(SalePrice)Í∞Ä right-skewedÎêòÏñ¥ ÏûàÏäµÎãàÎã§.
- ÌÉÄÍ≤üÏóê ÎåÄÌï¥ Î°úÍ∑∏ Î≥ÄÌôòÏùÑ Ìï¥ Î¥ÖÏãúÎã§.
  - ÎåÄÏπ≠Ïóê Í∞ÄÍπùÍ≤å Î∂ÑÌè¨Í∞Ä Î≥ÄÌïú Í≤ÉÏùÑ ÌôïÏù∏Ìï† Ïàò ÏûàÏäµÎãàÎã§.

```python
import numpy as np

fig, ax = plt.subplots(1, 2, figsize=(12, 5))
sns.histplot(df[target], ax=ax[0])
sns.histplot(np.log1p(df[target]), ax=ax[1])
```

### TransformedTargetRegressorÏùÑ ÏÇ¨Ïö©Ìï¥ Î¥ÖÏãúÎã§.
- `scikit-learn`ÏóêÏÑúÎäî ÌÉÄÍ≤üÏùÑ Î≥ÄÌôòÌïòÏó¨ ÌöåÍ∑ÄÏòàÏ∏°ÌïòÎäî `TransformedTargetRegressor` ÌÅ¥ÎûòÏä§Î•º Ï†úÍ≥µÌï©ÎãàÎã§.

```python
from sklearn.model_selection import train_test_split

df = df.dropna(subset=[target])
train, val = train_test_split(df, test_size=0.2, random_state=42)

X_train = train[features]
y_train = train[target]
X_val = val[features]
y_val = val[target]
```

```python
from category_encoders import OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.compose import TransformedTargetRegressor

pipe = make_pipeline(OrdinalEncoder(), SimpleImputer(), LinearRegression())
pipe.fit(X_train, y_train)
pipe.score(X_val, y_val)
```

### Í∏∞Ï°¥ Regression PipelineÏóê `TransformedTargetRegressor`ÏùÑ ÏîåÏõå Í∞ÑÌé∏ÌïòÍ≤å Íµ¨ÌòÑÌï† Ïàò ÏûàÏäµÎãàÎã§.

```python
trans_target_reg = TransformedTargetRegressor(
    regressor=pipe, func=np.log1p, inverse_func=np.expm1
)

trans_target_reg.fit(X_train, y_train)
trans_target_reg.score(X_val, y_val)
```
