## 1. 머신러닝 문제의 특성
### 내 문제를 풀기에 ML이 적합할까요?
- ML 기술이 우리 삶의 곳곳에 스며들고 있습니다. 많은 사람들이 ML을 `모든 문제를 해결해 주는 만능 열쇠`라고 착각합니다.
- 이는 사실이 아닙니다. ML 문제는 많은 데이터와 계산 자원을 필요로 하며, 복잡한 feature engineering 및 실험 과정이 수반됩니다.
- 따라서, ML을 사용해 문제를 해결하고자 한다면 아래의 조건을 충족하는지를 확인해야 합니다.

#### 1. 조건에 맞는 데이터가 있습니다.
- 데이터는 ML을 구성하는 필수적인 요소입니다. ML 문제를 풀기 위해서는 갖고 있는 데이터가 여러 조건에 부합하는지 확인해야 합니다.
    1. **충분한 양의 데이터**: '충분함'의 기준은 문제마다 다르지만, 단순 회귀 문제의 경우 수백 ~ 수천 개 이상, 모델이 복잡해질 경우 만 단위 이상의 데이터가 요구됩니다. 데이터가 많을수록 모델의 일반화 성능은 올라갑니다.
    2. **풀고자 하는 문제상황을 잘 반영하는 데이터**: ML 모델은 학습 때 본 적이 없는 형태의 데이터가 입력되면 이상한 출력값을 낼 수 있습니다. 본인이 갖고 있는 데이터가 풀고자 하는 문제상황을 예외 없이 잘 반영하는지, 혹시 특정 집단에 치우친 데이터는 아닌지 확인해야 합니다.
    3. **타겟과의 상관 관계가 있을 것으로 추정되는 데이터**: 아무리 항목이 많은 데이터라 할지라도 타겟과의 상관관계가 전혀 없다면 ML 모델은 학습되기 어렵습니다. ML 문제를 풀기 전, 본인이 갖고 있는 데이터의 몇몇 항목들이 타겟과 강력한 상관관계를 갖고 있을 것이라는 가설이 필요합니다.

#### 2. Non-ML 방법으로는 풀기 어렵거나, 현재의 Non-ML 방법이 더 이상 성능 향상의 여지가 보이지 않습니다.
- 흔히 휴리스틱 알고리즘(간단하고 빠르게 문제에 적용할 수 있는 솔루션)으로 대표되는 Non-ML 방법들은 대개 적은 데이터와 자원만을 활용하여 구현될 수 있습니다.
- ML 시스템의 설계에는 수많은 실험 및 검증이 필요하며, 대개 ML 기반 서비스가 생성한 의사결정 결과는 설명할 수 없는 형태로 주어지기 때문에 결과의 이해에도 추가 자원이 필요합니다.
- 따라서 ML을 도입하기 이전 동일한 문제를 푸는 효율적인 Non-ML 방법에 대해 선행조사를 하는 과정이 반드시 필요합니다.

#### 3. 장/단기적으로 ML 시스템을 지속하기 위한 비용을 부담할 수 있습니다.
- 실제로 많은 서비스들은 ML 시스템을 관리하기 위해 기존의 서비스와 분리된 별도의 컴퓨팅 자원을 활용하며, 이를 지속적으로 관리하는 인적 자원을 투입합니다.
- 비슷한 규모와 주제의 문제를 해결하는 타 ML 서비스의 소요 자원 현황을 선행 조사하고 본인의 가용 자원을 점검하는 과정이 필요합니다.


#### 4. ML 모델은 예측값이 액션으로 연결될 때 가치를 가집니다.
- 현 상황의 단순 분석을 위해서는 **ML 모델보다는 통계적인 기법을 활용**하는 편이 좋습니다.
- ML 모델의 예측값이 본인이 제공하고자 하는 서비스에서의 어떤 행위로 이어지는지를 확인해야 합니다.  
    ex)
    - 추천(분류) 문제에서, 사용자가 클릭할 확률 예측값이 0.4 이상일 때만 해당 아이템을 피드에 제시합니다.
    - 매출 예측(회귀) 문제에서, 매출 예측값이 목표값보다 낮을 때 적절한 조치를 취합니다.

## 3. 머신러닝 문제 정의하기
### 분류 문제? 회귀 문제?
- 우선 풀고자 하는 문제를 분류 문제와 회귀 문제 중 어느 것으로 정의할지 결정해야 합니다.
    - 분류 문제와 회귀 문제는, 사용하고자 하는 타겟의 유형에 따라 구분되는 것이 **아닙니다**.
- 회귀 문제로 정의되는 많은 ML문제의 경우, 예측값이 일정 threshold를 넘는지 여부에 따라 취하는 액션이 달라집니다.
    - 예를 들어, 상품 추천 문제의 경우 추천 예측 점수가 일정 점수 이상일 때만 사용자에게 아이템을 제시합니다.
    - 주가 예측 등의 문제에서, 내일의 종가 값 자체보다 내일 종가가 오늘 종가에 비해 오를지 여부가 더 중요한 경우가 있습니다.
    - 회귀 문제로 정의할 경우 모델이 우리가 정하는 threshold와 관련된 정보를 전혀 알지 못하기 때문에, 이러한 경우 분류 문제로 해결하는 것이 좋습니다.

### 트리 모델을 사용해 분류기를 학습시켜 봅시다.
- 학습 데이터와 테스트 데이터를 분리합니다.
- 예측 타겟은 `is_green` 변수이며, 0 또는 1의 값을 가집니다.


```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.2, random_state=42)
```

```python
features = ["manufacturer", "class", "ttype", "ftype"] + [
    "esize",
    "n_cylinder",
    "fcons_mpg",
    "smog_rating",
]
target = "is_green"
X_train, y_train = train[features], train[target]
X_test, y_test = test[features], test[target]
```

### `pipeline`을 만들어 `DecisionTreeClassifier`을 학습합니다.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import make_pipeline
from category_encoders import OrdinalEncoder

pipe = make_pipeline(
    OrdinalEncoder(cols=["manufacturer", "class", "ttype", "ftype"]),
    DecisionTreeClassifier(max_depth=6, random_state=42),
)
pipe.fit(X_train, y_train)
print("검증 정확도", pipe.score(X_test, y_test))
```

### 학습과 검증 데이터셋을 잘 분리했는데도 검증 정확도가 1이 나왔습니다. 왜 그럴까요?
- 우선 어떤 트리가 만들어졌는지 그려봅시다.

```python
import graphviz
from sklearn.tree import export_graphviz

tree = pipe.named_steps["decisiontreeclassifier"]

dot_data = export_graphviz(
    tree,
    feature_names=X_train.columns,
    class_names=["False", "True"],
    filled=True,
    proportion=True,
)

graphviz.Source(dot_data)
```

- smog_rating이 5.5 이하이면 False, 그렇지 않으면 True로 분류합니다.
  - 이건 우리가 원하는 모델링 결과가 아닙니다.
  - 왜 이런 문제가 발생했을까요?

## 정보 누수(Data Leakage)

- 데이터에 대한 충분한 이해 없이 기계적으로 모델링할 경우 종종 Data Leakage 문제에 맞닥뜨리게 됩니다.
- 별다른 작업을 하지 않았음에도 100%에 가까운 일반화 성능을 얻은 경우 Data Leakage를 의심해 볼 필요가 있습니다.
- Data Leakage는 두 가지 경우로 나눌 수 있습니다.

### 1. 학습 데이터가 예측 시 못 쓰는 피쳐를 반영하는 경우(Column axis)
- 위 Case Study의 경우를 봅시다. smog_rating은 우리가 예측하고자 하는 값과 관련된 것이지, 실제 예측 시 사용할 수 있는 데이터는 아닐 것입니다.
- 이런 정보가 학습 시 피쳐로 포함된다면, 학습된 모델을 실제 상황에서 적용하기 어려울 것입니다.
- 이 유형의 Data Leakage의 발생 여부는 **모델의 설계 및 사용 목적을 기준으로 판단합니다.**
  - 모델링 시 `이 모델이 실제 사용되어 미지의 데이터를 가지고 예측할 때, 우리가 사용할 수 있는 피쳐 이상의 피쳐를 포함하고 있는가?`를 지속적으로 점검해야 합니다.
- **Solution**
    - 모델이 실제 사용될 때 어떤 데이터만을 가지고 타겟을 예측해야 하는지를 분명히 파악한다.
    - corr(상관계수)이 너무 높은 피쳐를 drop한다.
  
### 2. 학습 과정에서 평가 데이터의 정보가 활용되는 경우(Row axis)
- 학습 데이터와 검증 / 평가 데이터를 완전히 분리하지 않으면, 모델이 답을 미리 보고 문제를 푸는 꼴이 됩니다.
- 이 경우 비정상적으로 높은 일반화 성능을 얻을 수 있습니다.
- **Solution**
    - 학습 / 검증 / 평가 데이터를 **나눈 후 scaling 등을 진행**한다.
    - scaling / encoding 과정에 유의한다.
      - fit_transform은 학습 데이터에 대해서만 사용
      - 검증 데이터에 대해서는 fit_transform이 아닌 transform을 사용
      - 왜 그래야 할까요?
    - pipelining 사용해서 코딩 실수에 의한 오류를 줄인다.


### Data Leakage를 발생시킨 `smog_rating` 피쳐를 빼고 다시 학습을 진행해 봅시다.

```python
X_train = X_train.drop(["smog_rating"], axis=1)
X_test = X_test.drop(["smog_rating"], axis=1)
```

```python
pipe = make_pipeline(
    OrdinalEncoder(cols=["manufacturer", "class", "ttype", "ftype"]),
    DecisionTreeClassifier(max_depth=6, random_state=42),
)
pipe.fit(X_train, y_train)
print("검증 정확도", pipe.score(X_test, y_test))
```

```python
tree = pipe.named_steps["decisiontreeclassifier"]

dot_data = export_graphviz(
    tree,
    feature_names=X_train.columns,
    class_names=["False", "True"],
    filled=True,
    proportion=True,
)

graphviz.Source(dot_data)
```

## 평가지표 설정
- 문제의 유형에 맞는 평가지표를 선택하여 모델의 성능을 평가해야 합니다.
  - 분류 평가지표와 회귀 평가지표를 혼동해선 안 됩니다.
  - 하나의 평가지표만을 사용해 모델 성능을 검증하는 경우는 거의 없습니다. 여러 평가지표를 함께 살펴보며 모델의 예측 양상을 파악합니다.
<br></br>

- 대표적인 회귀 평가지표는 아래와 같습니다.
  - $MSE$: 미분 가능하여 비용 함수로 가장 흔하게 사용됩니다. 하지만 제곱을 취하여 이상치에 민감합니다.
  - $RMSE$: $MSE$에 루트를 씌워, 이상치에 상대적으로 덜 민감합니다.
  - $MAE$: 절대오차의 평균으로, 직관적으로 이해하기 쉬워 선호됩니다.
  - $R^2$: 모델이 데이터들을 얼마나 잘 설명하는지를 보여줍니다.
<br></br>

- 대표적인 분류 평가지표는 아래와 같습니다.
  - $Accuracy=\frac{TP+TN}{TP+TN+FP+FN}$
  - $Precision=\frac{TP}{TP+FP}$
  - $Recall=\frac{TP}{TP+FN}$
  - $F1=\frac{2}{1/Precision + 1/Recall}$
  - $ROC-AUC$

  - 타겟분포가 불균형할 때는 accuracy는 적절한 평가지표가 아닙니다. 이 외에 다른 평가지표로 모델의 성능을 평가해야합니다.

### 분류 문제에서는 Confusion Matrix를 계산하여 직관적으로 모델의 예측 결과를 확인할 수 있습니다.

```python
from sklearn.metrics import plot_confusion_matrix

plot_confusion_matrix(pipe, X_test, y_test, cmap=plt.cm.Reds)
plt.title(f"Confusion Matrix (n = {len(y_test)})", fontsize=15)
```

### Classification Report를 통해 각 class에 대한 성능 평가 결과를 확인할 수 있습니다.

```python
from sklearn.metrics import classification_report

y_pred = pipe.predict(X_test)
print(classification_report(y_test, y_pred))
```

### ROC 곡선과 AUC 스코어를 통해 모델이 예측한 값(확률, Logit 등)에 대한 성능을 확인할 수 있습니다.
- ROC 곡선은 FPR 변화에 따른 TPR의 변화의 자취를 찍은 그래프입니다.
- AUC 스코어는 ROC 곡선 아랫부분의 면적이며, 값이 클 수록 모델은 True 데이터와 False 데이터를 잘 구분합니다.

```python
from sklearn.metrics import roc_curve, roc_auc_score

y_pred_proba = pipe.predict_proba(X_test)[:, 1]
print("AUC:", roc_auc_score(y_test, y_pred_proba))

baseline_fpr, baseline_tpr, _ = roc_curve(y_test, [0] * len(y_test))
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

plt.plot(baseline_fpr, baseline_tpr, linestyle="--", color="green", label="Baseline")
plt.plot(fpr, tpr, color="blue", label="Model")
plt.scatter(fpr, tpr, color="red")
plt.legend()
plt.title("ROC curve")
plt.xlabel("FPR")
plt.ylabel("TPR")
```

## 타겟의 분포가 불균형일 때
- 많은 ML 모델은 타겟의 분포가 불균형할 때 데이터를 잘 학습하지 못합니다.
- 타겟 분포가 불균형일 때 모델이 데이터를 잘 학습하도록 하는 몇 가지 테크닉이 있습니다.


### 분류 문제
- 클래스 비율이 크게 차이날 경우, 학습 과정에서 모델이 소수 클래스의 정보를 충분히 학습하기 어렵습니다.
  - 이진 분류 문제에서, 일반적으로 한 클래스의 비율이 80%를 넘으면 불균형 문제가 있다고 진단합니다. 한 클래스의 비율이 99%를 넘으면 심각한 불균형이 존재한다고 진단합니다.
- 대표적으로 세 가지의 해결 방법이 있습니다.
  
1. **`class_weight`**
   - 각 클래스의 데이터에 대해 서로 다른 가중치를 부여해 학습합니다.
   - 소수 클래스의 데이터에 더 큰 가중치를 부여하여 불균형 문제를 해소할 수 있습니다.
<br></br>

2. **`sampling`**
   - oversampling : 소수 클래스의 데이터로부터 인공 데이터를 생성하여, 소수 클래스의 크기를 다수 클래스의 크기와 비슷하게 키웁니다.
   - undersampling : 다수 클래스의 데이터를 모두 학습에 사용하지 않고, 소수 클래스의 수만큼만 샘플링하여 사용합니다.

- `n234`에서 더 자세히 다뤄보겠습니다. 

### Case Study에서 불균형 문제를 해결해 봅시다.
- `class weight`를 주는 방식으로 불균형 문제를 해결해 보겠습니다.
- `scikit-learn` 등 대다수의 ML 라이브러리에서 모델 학습 시 `class_weight` 옵션을 제공합니다.

```python
pipe = make_pipeline(
    OrdinalEncoder(cols=["manufacturer", "class", "ttype", "ftype"]),
    DecisionTreeClassifier(max_depth=6, random_state=42, class_weight="balanced"),
)
pipe.fit(X_train, y_train)
print("검증 정확도", pipe.score(X_test, y_test))
```

```python
plot_confusion_matrix(pipe, X_test, y_test, cmap=plt.cm.Reds)
plt.title(f"Confusion Matrix (n = {len(y_test)})", fontsize=15)
```

```python
y_pred = pipe.predict(X_test)
print(classification_report(y_test, y_pred))
```

- 소수 클래스(True)의 F1 score가 개선되었습니다.

### 회귀 문제 - 분포 변환
- 회귀 문제에서는 타겟의 분포가 비대칭 형태인지 확인해 보세요.
- 선형 회귀 모델은 특성 변수들에 따른 타겟변수의 분포가 정규분포 형태일때 좋은 성능을 보입니다. 
  - 특히 타겟변수가 왜곡된 형태의 분포(skewed)일 경우 예측 성능에 부정적인 영향을 미칩니다.

<img src="https://ifh.cc/g/oOrSko.png" />


- 타겟이 right-skewed 상태라면 로그 변환을 사용하여 비대칭성을 완화할 수 있습니다.

<img src="https://ifh.cc/g/f5Z5Hp.png" width="600" />


- [log1p](https://numpy.org/doc/stable/reference/generated/numpy.log1p.html): `ln(1 + x)`
- [expm1](https://numpy.org/doc/stable/reference/generated/numpy.expm1.html#numpy.expm1): `exp(x) - 1`, the inverse of log1p.

### 🏠 Case Study: House Price Prediction

```python
df = pd.read_csv(
    "https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/house-prices/house_prices_train.csv"
)
```

```python
# 예측하고자 하는 컬럼은 SalePrice입니다. 분포를 확인해 볼까요?
features = [
    "BldgType",
    "CentralAir",
    "FullBath",
    "GrLivArea",
    "LotArea",
    "OverallCond",
    "OverallQual",
    "TotRmsAbvGrd",
    "YearBuilt",
    "YrSold",
]
target = "SalePrice"
sns.displot(df[target])
```

- 타겟(SalePrice)가 right-skewed되어 있습니다.
- 타겟에 대해 로그 변환을 해 봅시다.
  - 대칭에 가깝게 분포가 변한 것을 확인할 수 있습니다.

```python
import numpy as np

fig, ax = plt.subplots(1, 2, figsize=(12, 5))
sns.histplot(df[target], ax=ax[0])
sns.histplot(np.log1p(df[target]), ax=ax[1])
```

### TransformedTargetRegressor을 사용해 봅시다.
- `scikit-learn`에서는 타겟을 변환하여 회귀예측하는 `TransformedTargetRegressor` 클래스를 제공합니다.

```python
from sklearn.model_selection import train_test_split

df = df.dropna(subset=[target])
train, val = train_test_split(df, test_size=0.2, random_state=42)

X_train = train[features]
y_train = train[target]
X_val = val[features]
y_val = val[target]
```

```python
from category_encoders import OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.compose import TransformedTargetRegressor

pipe = make_pipeline(OrdinalEncoder(), SimpleImputer(), LinearRegression())
pipe.fit(X_train, y_train)
pipe.score(X_val, y_val)
```

### 기존 Regression Pipeline에 `TransformedTargetRegressor`을 씌워 간편하게 구현할 수 있습니다.

```python
trans_target_reg = TransformedTargetRegressor(
    regressor=pipe, func=np.log1p, inverse_func=np.expm1
)

trans_target_reg.fit(X_train, y_train)
trans_target_reg.score(X_val, y_val)
```
