## 2.1 ì„ í˜•íšŒê·€ëª¨ë¸

- ì‚°ì ë„(scatterplot)ì— ê°€ì¥ ì˜ ë§ëŠ”(best fit) ì§ì„ ì„ ê·¸ë ¤ì£¼ë©´ ê·¸ê²ƒì´ íšŒê·€ ì˜ˆì¸¡ëª¨ë¸ì´ ë©ë‹ˆë‹¤.
- ê°€ì¥ ì˜ ë§ëŠ” íšŒê·€ì„ ì„ ì–´ë–»ê²Œ êµ¬í•  ìˆ˜ ìˆì„ê¹Œìš”?

ë‹¨ìˆœí•˜ê²Œ ìƒê°í•´ë´…ì‹œë‹¤.  
- ëª¨ë¸ì˜ ì˜ˆì¸¡ê°’ê³¼ ì‹¤ì œê°’ì˜ ì°¨ì´ê°€ 0ì´ë¼ë©´ ëª¨ë¸ì€ ì‹¤ì œê°’ì„ ì •í™•í•˜ê²Œ ì˜ˆì¸¡í•˜ëŠ” ìµœê³ ì˜ ëª¨ë¸ì¼ ê²ƒì…ë‹ˆë‹¤.
- `(ì‹¤ì œê°’) - (ëª¨ë¸ì˜ ì˜ˆì¸¡ê°’)`ì„ **ì”ì°¨(residual)**ë¼ê³  í•˜ë©° ì”ì°¨ë¥¼ ìµœì†Œë¡œ í•˜ëŠ” ì„ ì„ ì°¾ìœ¼ë©´ ë©ë‹ˆë‹¤.  
    - ì‹¤ì œê°’ : $ y $
    - ë‹¨ìˆœì„ í˜•íšŒê·€ëª¨ë¸(ì˜ˆì¸¡ê°’) : ${\hat y} = \hat{\alpha}x + \hat{\beta} $
    - ì”ì°¨ : $ y - (\hat{\alpha}x + \hat{\beta}) $  
    cf) ì˜¤ì°¨(error)ëŠ” ëª¨ì§‘ë‹¨ì—ì„œì˜ ì˜ˆì¸¡ê°’ê³¼ ì‹¤ì œê°’ì˜ ì°¨ì´ë¥¼ ë§í•©ë‹ˆë‹¤. í•˜ì§€ë§Œ í†µìƒì ìœ¼ë¡œëŠ” ê°™ì€ ì˜ë¯¸ë¡œ ì‚¬ìš©ë©ë‹ˆë‹¤.
    

<br></br>

ìµœì ì˜ ì„ í˜•íšŒê·€ì„ ì€ ì”ì°¨ ì œê³±ë“¤ì˜ í•©ì¸ RSS(residual sum of squares)ë¥¼ ìµœì†Œë¡œ í•˜ëŠ” ì§ì„ ì…ë‹ˆë‹¤.

$$ RSS = \sum_{i=1}^{N} {\hat{e_i}}^2 =  \sum_{i=1}^{N} {(y_i - (\hat{\alpha}x_i + \hat{\beta}))}^2 $$

RSSëŠ” ì„ í˜•íšŒê·€ëª¨ë¸ì˜ **ë¹„ìš©í•¨ìˆ˜**ì…ë‹ˆë‹¤. ëª¨ë¸ì„ í•™ìŠµí•œë‹¤ëŠ” ê²ƒì€ ì´ ë¹„ìš©í•¨ìˆ˜ë¥¼ ìµœì†Œë¡œ í•˜ë„ë¡ í•˜ëŠ” íŒŒë¼ë¯¸í„° $\alpha$ì™€ $\beta$ë¥¼ ì°¾ëŠ” ê²ƒì…ë‹ˆë‹¤.  
$\alpha$, $\beta$ëŠ” ë¹„ìš©í•¨ìˆ˜ë¥¼ ë¯¸ë¶„í•˜ì—¬ ë‚˜ì˜¨ ì•„ë˜ì™€ ê°™ì€ ì‹ì„ í†µí•´ êµ¬í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì•„ë˜ ì‹ì„ ë„ì¶œí•˜ëŠ” ê³¼ì •ì€ ì°¸ê³ ë¬¸ì„œë¥¼ ì°¸ì¡°í•´ì£¼ì„¸ìš”.

$ \hat{\alpha} = \frac{S_{xy}}{S_{xx}} $  

$ \hat{\beta} = \bar{y} - \hat{\alpha}\bar{x}$


$ S_{xy}= {\sum_{i=1}^{N}(x_i-\bar{x})(y_i-\bar{y})}$

$ S_{xx} = {\sum_{i=1}^{N}(x_i-\bar{x})^2}$


ì´ëŸ° ì‹ìœ¼ë¡œ ì”ì°¨ ì œê³±ì˜ í•©ì„ ìµœëŒ€í•œ ì‘ê²Œ, ì¦‰ **ìµœì†Œ**ë¡œ ë§Œë“¤ì–´ ì£¼ëŠ” ë°©ë²•ì„ **ìµœì†Œì œê³±ë²•**, ì˜ì–´ë¡œëŠ” **Ordinary Least Square**, ì¤„ì—¬ì„œ **OLS**ë¼ê³  ë¶€ë¦…ë‹ˆë‹¤.

### ë‹¤í•­ì„ í˜•íšŒê·€(Polynomial Linear Reregrssion)
- ë‹¤í•­ì„ í˜•íšŒê·€ëŠ” 2ì°¨í•­ ì´ìƒì˜ ë…ë¦½ë³€ìˆ˜ë¡œ ì¢…ì†ë³€ìˆ˜ë¥¼ ì„¤ëª…í•˜ëŠ” ì„ í˜•íšŒê·€ì…ë‹ˆë‹¤.
- ë…ë¦½ë³€ìˆ˜ê°€ 2ê°œì¸ ë‹¤í•­ì„ í˜•íšŒê·€ëŠ” ì•„ë˜ì™€ ê°™ì€ ì‹ìœ¼ë¡œ ë‚˜íƒ€ë‚¼ ìˆ˜ ìˆìŠµë‹ˆë‹¤.(degree=2)

  $ y = {\alpha_{1}}x_1 + {\alpha_{2}}x_2 + {\alpha_{1}}x_1^2 + {\alpha_{2}}x_2^2 + {\alpha_{1}}{x_1}{x_2} + \beta $
<br></br>

> ë‹¤í•­ì„ í˜•íšŒê·€ë„ ì—­ì‹œ ì„ í˜•íšŒê·€ì…ë‹ˆë‹¤. ì™œì¼ê¹Œìš”? ì„ í˜•íšŒê·€ì—ì„œ ë§í•˜ëŠ” `ì„ í˜•ì„±`ì´ë€ ë¬´ì—‡ì¼ê¹Œìš”?

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures

# PolynomialFeaturesë¥¼ ì‚¬ìš©í•˜ë©´ ë‹¤í•­íšŒê·€ë¥¼ ë§Œë“¤ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

def PolynomialRegression(degree=2, **kwargs):
    return make_pipeline(PolynomialFeatures(degree), 
                         LinearRegression(**kwargs))  # make_pipelineì€ ë‘ í•¨ìˆ˜ë¥¼ ë™ì‹œì— ë³¼ ìˆ˜ ìˆìŒ


features = ['GrLivArea', 'OverallQual']
target = 'SalePrice'

X = df[features]
y = df[target]

# 2ì°¨í•­ì˜ ë‹¤í•­ì„ í˜•íšŒê·€ë¥¼ ë§Œë“¤ì–´ë³´ê² ìŠµë‹ˆë‹¤.

poly_ols = PolynomialRegression(degree=2)
poly_ols.fit(X, y)
```

```python
# ë³´ê°„
interpolate_x_p = [[4000, 5]]
interpolate_y_p = poly_ols.predict(interpolate_x_p)[0]

# ì™¸ì‚½
extrapolate_x_p = [[6000, 10]]
extrapolate_y_p = poly_ols.predict(extrapolate_x_p)[0]

print(f'GrLivAreaê°€ 4000ì´ê³  OverallQualì´ 5 ê²½ìš° ì˜ˆì¸¡ê°’ì€ ${round(interpolate_y_p,2)}ì…ë‹ˆë‹¤.')
print(f'GrLivAreaê°€ 6000ì´ê³  OverallQualì´ 10 ê²½ìš° ì˜ˆì¸¡ê°’ì€ ${round(extrapolate_y_p,2)}ì…ë‹ˆë‹¤.')
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['GrLivArea'], y=df['SalePrice'], hue=df['OverallQual'], size=df['OverallQual'], alpha=0.6)
sns.scatterplot(x = [4000, 6000], y=[interpolate_y_p, extrapolate_y_p], color='red')


plt.text(4000, interpolate_y_p, 'interpolate value\n of 4000 & 5')
plt.text(6000, extrapolate_y_p, 'extrapolate value\n of 6000 & 10')

plt.show()
```

* `MSE (Mean Squared Error)` = $\frac{1}{n}\sum_{i=1}^{n}(y_{i} - \hat{y_{i}})^{2}$
    - MSEëŠ” ì œê³±ì˜¤ì°¨ì˜ í‰ê· ì…ë‹ˆë‹¤.
    - MSEëŠ” ë¯¸ë¶„ì´ ê°€ëŠ¥í•˜ì—¬ ë¹„ìš© í•¨ìˆ˜ë¡œ ê°€ì¥ í”í•˜ê²Œ ì‚¬ìš©ë©ë‹ˆë‹¤.

* `RMSE (Root Mean Squared Error)` = $\sqrt{MSE}$
    - RMSEëŠ” MSEì— ë£¨íŠ¸ë¥¼ ì”Œì›Œ ìƒëŒ€ì ìœ¼ë¡œ ëœ ë¯¼ê°í•˜ê²Œ ë©ë‹ˆë‹¤.

    - ê°’ì˜ í¬ê¸°ê°€ ì‹¤ì œê°’ê³¼ ë¹„ìŠ·í•´ì§€ê¸° ë•Œë¬¸ì— MSEì— ë¹„í•´ ì´í•´í•˜ê¸° ì‰½ìŠµë‹ˆë‹¤.
<br></br>

* `MAE (Mean absolute error)` = $\frac{1}{n}\sum_{i=1}^{n}\left | y_{i} - \hat{y_{i}} \right |$
    - MAEëŠ” ì ˆëŒ€ì˜¤ì°¨ì˜ í‰ê· ìœ¼ë¡œ ì§ê´€ì ìœ¼ë¡œ ì´í•´í•˜ê¸° ì‰½ìŠµë‹ˆë‹¤.

<br></br>

* `R-squared (Coefficient of determination)` = $1 - \frac{\sum_{i=1}^{n}(y_{i} - \hat{y_{i}})^{2}}{\sum_{i=1}^{n}(y_{i} - \bar{y_{i}})^{2}} = 1 - \frac{SSE}{SST} = \frac {SSR}{SST}$
    - $R^2$ (ê²°ì •ê³„ìˆ˜)ëŠ” ëª¨ë¸ì˜ ì„¤ëª…ë ¥ì„ ë‚˜íƒ€ëƒ…ë‹ˆë‹¤.

<br></br>

- ì°¸ê³ 
    - SSE(Sum of Squares `Error`, ê´€ì¸¡ì¹˜ì™€ ì˜ˆì¸¡ì¹˜ ì°¨ì´): $\sum_{i=1}^{n}(y_{i} - \hat{y_{i}})^{2}$
    - SSR(Sum of Squares due to `Regression`, ì˜ˆì¸¡ì¹˜ì™€ í‰ê·  ì°¨ì´): $\sum_{i=1}^{n}(\hat{y_{i}} - \bar{y_{i}})^{2}$
    - SST(Sum of Squares `Total`, ê´€ì¸¡ì¹˜ì™€ í‰ê·  ì°¨ì´): $\sum_{i=1}^{n}(y_{i} - \bar{y_{i}})^{2}$ , SSE + SSR

```python
# scikit-learnì„ í†µí•´ ì†ì‰½ê²Œ í‰ê°€ì§€í‘œë¥¼ ì‚¬ìš©í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
# 4ê°€ì§€ ëª¨ë¸ì˜ ì„±ëŠ¥ì„ í‰ê°€í•´ë³´ê² ìŠµë‹ˆë‹¤.

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
# ì‹¤ì œê°’
y_real = df['SalePrice']

# ê¸°ì¤€ ëª¨ë¸
y_base = [predict]*len(df)

# ë‹¨ìˆœì„ í˜•íšŒê·€
feature = ['GrLivArea']
X_simple = df[feature]
y_simple = simple_ols.predict(X_simple)

# ë‹¤ì¤‘ì„ í˜•íšŒê·€
features = ['GrLivArea', 'OverallQual']
X_multiple = df[features]
y_multiple = multiple_ols.predict(X_multiple)

# ë‹¤í•­ì„ í˜•íšŒê·€
y_poly = poly_ols.predict(X_multiple)
```

```python
# mse, rmse, mae, r2ë¥¼ ì¶œë ¥í•˜ëŠ” í•¨ìˆ˜ì…ë‹ˆë‹¤.

def eval_models(y_pred, y_real=y_real) :
    mse = mean_squared_error(y_real, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_real, y_pred)
    r2 = r2_score(y_real, y_pred)

    return mse, rmse, mae, r2
```

```python
base_mse, base_rmse, base_mae, base_r2 = eval_models(y_base)
simple_mse, simple_rmse, simple_mae, simple_r2 = eval_models(y_simple)
multiple_mse, multiple_rmse, multiple_mae, multiple_r2 = eval_models(y_multiple)
poly_mse, poly_rmse, poly_mae, poly_r2 = eval_models(y_poly)
```

```python
comparison_metrics = pd.DataFrame(index=['mse', 'rmse', 'mae', 'r2'], columns=['Base','Simple', 'Multiple', 'Polynomial'])
comparison_metrics['Base'] = [base_mse, base_rmse, base_mae, base_r2 ]
comparison_metrics['Simple'] = [simple_mse, simple_rmse, simple_mae, simple_r2]
comparison_metrics['Multiple'] = [multiple_mse, multiple_rmse, multiple_mae, multiple_r2]
comparison_metrics['Polynomial'] = [poly_mse, poly_rmse, poly_mae, poly_r2]
comparison_metrics
```
